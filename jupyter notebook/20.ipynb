金額欠損の補完
商品名が同じものは全て同じ値段に変更（データ期間内に値段変更がない前提のため）

```python
flg_is_null = uriage_data["item_price"].isnull()                                                   #item_priceが欠損してる行を格納
for trg in list(uriage_data.loc[flg_is_null, "item_name"].unique()):                               #欠損してる行のitem_nameを取得してループ
    price = uriage_data.loc[(~flg_is_null) & (uriage_data["item_name"] == trg),"item_price"].max() #欠損してない行でitem_naemが一致したものがある時、その金額を一意にして取得 （このままだとコピー値の変更による警告なので以下で回避）
    for index,flg in enumerate(flg_is_null):
        if flg and uriage_data.loc[index,"item_name"] == trg:
            uriage_data.loc[index,"item_price"] = price
#     uriage_data["item_price"].loc[(flg_is_null) & (uriage_data["item_name"]==trg)] = price         #取得した金額を欠損のある行で、item_nameが一致してるところに入れる
uriage_data
```

16.日付の揺れの補正 揺れを確認する

日付ではない数字がある。<br>
2018/01/04<br>
42782<br>
原因は取り込み元のデータ（Excel）<br>
2018/01/04  （フォント違い）<br>
2017年2月16日<br>
2018/01/07<br><br>
数字は1900/01/01をスタートにした経過日数（シリアル値）で、エクセルと比べて-2日ずれて変換されている<br>
１．エクセルの最小日付は1900/01/01でシリアル値が1だが、pythonでは0と認識する<br>
２．エクセルではうるう年でない日付もシリアル値で有効になるが、pythonでは無効となるためカウントされない

```python
flg_is_serial = kokyaku_data["登録日"].astype("str").str.isdigit() #strにキャストして、数値の行を取り出す
flg_is_serial.sum()
```

```python
#数値の行は経過時間として1900/01/01をプラス
fromSerial = pd.to_timedelta(kokyaku_data.loc[flg_is_serial, "登録日"].astype("float")-2,unit="D") + pd.to_datetime("1900/01/01") #-2はエクセルのシリアル値の対応
fromSerial
```

```python
#その他の日付も書式統一
fromString = pd.to_datetime(kokyaku_data.loc[~flg_is_serial, "登録日"])
fromString
```

```python
kokyaku_data["登録日"] = pd.concat([fromSerial, fromString]) #日付を結合して変更
kokyaku_data
```

揺れが消えたか検証１ データ件数でチェック

```python
kokyaku_data["登録年月"] = kokyaku_data["登録日"].dt.strftime("%Y%m") # 登録日を年月に変換
rslt = kokyaku_data.count()["顧客名"] # 合計数を出す
print(rslt)
rslt = kokyaku_data.groupby("登録年月").count()["顧客名"] # groupbyで各顧客名の件数を取得
print(rslt)
print(len(kokyaku_data)) #データ全体の件数と合ってるか
```
